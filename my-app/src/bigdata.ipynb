```python
import csv
import math
import matplotlib.pyplot as plt

f=open('gender.csv', 'r', encoding='cp949')
data = csv.reader(f)
m=[]
f=[]
size=[]
name=input('원하는 동 입력 : ')

for row in data :
  if name in row[0]:
    for i in range(3, 104):
      temp1 = int(row[i].replace(",",""))
      temp2 = int(row[i+103].replace(",",""))
      m.append(temp1)
      f.append(temp2)
      size.append(math.sqrt(temp1+temp2))
      #size.append(temp1+temp2)
    break

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=100)
plt.scatter(m,f,s=size, c=range(101), alpha=0.5, cmap='jet')#컬러맵 적용
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g')#추세선 추가
plt.xlabel('Male population')
plt.ylabel('Female population')
plt.show()
```

```python
#유임승차와 무임승차의 비율의 최댓값 찾기
import csv
file = open('subwayfee.csv', 'r', encoding='cp949')
data = csv.reader(file)
next(data) #표의 제목들의 다음 행부터 할 것이기 때문

mx=-1 #초기값을 양수보다 작게 하기 위해서 rate 값이 바로 갱신되도록
rate=0
mx_station = ''

for row in data:
  for i in range(4,8):
    row[i] = int(row[i])
  if(row[6]==0 and row[4]+row[6]<100000): #row[6]에 0이 있기 때문에, 0이 있어도 계속 진행하라는 뜻.
    continue
  rate = row[4]/ (row[4]+row[6]) #나눈것임
  if rate > mx :
    mx=rate
    mx_station = row[3] + ' ' + row[1]
  
print(mx_station, round(mx,2))
```

```python
#유임승차와 무임승차의 비율의 최댓값 찾기
import csv
file = open('subwayfee.csv', 'r', encoding='cp949')
data = csv.reader(file)
next(data) #표의 제목들의 다음 행부터 할 것이기 때문

mx=[-1]*4
mx_station = ['']*4
label = ['유임승차', '유임하차', '무임승차', '무임하차']
#print(mx, mx_station, label)

for row in data:
  for i in range(4,8):
    row[i] = int(row[i])#row의 첫째 둘째~ 등 값이 
    if(row[i] > mx[i-4]):
      mx[i-4] = row[i]
      mx_station[i-4] = row[3] + ' ' + row[1]
#처음 봤을 때, 가시적ㅇ

for i in range(4):
  print(label[i]+" : "+mx_station[i], mx[i])
```
