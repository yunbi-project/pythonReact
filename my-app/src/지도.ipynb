```python
from folium.map import Marker
import folium
import pandas as pd

#파일 읽기
df = pd.read_excel('/content/university_position_seoul.xlsx', engine='openpyxl')
df.set_index('name', inplace=True)

#서울 지도 만들기
seoul_map = folium.Map(location=[37.55,126.98], tiles='Stamen Terrain', zoom_start=12)

#대학교 위치 정보를 Marker로 표시
#zip:동일한 개수로 이루어진 자료형을 묶어주는 역할을 하는 함수
for name, lat, lng in zip (df.index, df.latitude, df.longitude):
  folium.CircleMarker([lat, lng],
                     radius=10,
                     color='brown',
                     fill=True,
                     fill_color='coral',
                     fill_opacity=0.7,
                     popup=name).add_to(seoul_map)

#지도를 HTML 파일로 저장하기
seoul_map.save('/content/seoul_univ.html')
```
