```python
from folium.map import Marker
import folium
import pandas as pd

#파일 읽기
# df = pd.read_excel('/content/university_position_seoul.xlsx', engine='openpyxl')
# df.set_index('name', inplace=True)
# 임의의 대학교 위치 데이터 생성
data = {
    'name': ['University A', 'University B', 'University C'],
    'latitude': [37.55, 37.56, 37.57],
    'longitude': [126.98, 126.99, 127.0]
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 서울 지도 만들기
seoul_map = folium.Map(location=[37.55, 126.98], tiles='Stamen Terrain', attr='Map data © OpenStreetMap contributors, Imagery © Stamen Design', zoom_start=12)

# 대학교 위치 정보를 Marker로 표시
for index, row in df.iterrows():
    folium.Marker([row['latitude'], row['longitude']], popup=index).add_to(seoul_map)

# 지도를 HTML 파일로 저장하기
seoul_map.save('seoul_univ.html')
```
